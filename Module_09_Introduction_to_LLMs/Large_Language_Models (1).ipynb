## What are LLMs?

LLMs are artificial neural networks that are trained on a large amount of text data. They are trained to predict the next tokens (words) when passed in some input tokens. For example when passed in the tokens "She is a computer" the model may predict "scientist" as the next token, because it has been trained with the similar (or the same) sentence previously.

LLMs can also be utilized to solve other tasks like question answering, sentiment analysis, summarization, etc. For example, if an LLM was passed in the input "Answer the question. Q: What is the capital of the United States? A:", it would try to complete the text with the most likely token, which is probably the answer to the question.

Some examples of popular LLMs are ChatGPT, LLaMA, Gemini and Claude.

## Prompting
A **prompt** is an input text given to an LLM with a goal to look for an appropriate output or **completion**. The example discussed in the previous section, "Answer the question. Q: What is the capital of the United States? A:" is a prompt to the LLM to look for an answer to the question.

When the LLM model was trained, it has most likely seen a lot of questions and answers in the form of "Q: \<QUESTION\> A: \<ANSWER\>" so the model learns to try to answer the question, when presented with text in that format.

Review these prompting tips: https://www.promptingguide.ai/introduction/tips

To create your own prompt:
- Start with a simple sentence or question, and then slowly add more details and context
- Use commands to instruct the model, such as "Write", "Classify", "Summarize", "Translate”
- Be specific and direct
- Focus primarily on saying what the model should do instead instead of what the model should not do (though sometimes, if the model repeatedly gets it wrong, in later interactions, you might need to direct it on what not to do.)


Two key things to keep in mind when using LLMs to support your knowledge development:

1) Experiment and iterate on your prompts. You should test your prompts with some examples and see their outputs; then refine your prompts to improve the generation quality over a few iterations.

2) If your taks is complex, give the LLMs contained modular sub-tasks to do and build on it as you move forward.

For more information on prompting you can look at the DAIR AI prompt guide: https://www.promptingguide.ai/

## Objective (Loss)

LLMs main goal is to predict the best next token after being passed in the previous tokens. Said another way, LLMs take in some input tokens (words) and output *probabilities* for the next token, where the "best" tokens will have the highest probability. When training a language model, the objective is to maximimize the probability of the best next token.

LLMs use a language modelling objective, using the following likelihood function:
$$L(U) = \sum_i\log P(u_i|u_{i−k}, \dots , u_{i−1}; \Theta)$$

This essentially says you want to maximize the probability of the next token $u_i$ based on the previous tokens $u_{i−k}, \dots , u_{i−1}$ and the model parameters $\Theta$.

When training an LLM, you take a large dataset of text (corpus), and you sample some text from it. You pass in a few tokens into the model and get the model to predict the next token, and you reward the model for giving a higher probability to the next token. For example, let's say the sentence "She is a computer scientist" in the corpus. You would feed the model "She is a computer " and then ask the model to predict the probabilities of the next token, and reward the model if the probability of "scientist" was higher.

## Decoding
The simple output of the LLM is the next token probabilities. The process of generating multiple tokens from the LLM to produce the entire output text is referred to as the **decoding** process. It is a sequential process where tokens are generated one at a time. The generation of each token is based on the previously generated tokens.

**Example:**

For decoding, the LLM starts by generating a probability value for each possible token in our vocabulary set. For example if we have a vocabulary of 6 tokens \[I, you, like, love, pizza, pasta\]. Let's say you start by feeding in the LLM a single token, "you". The LLM would generate a probability vector of length 6 e.g. \[.1, .1, .3, .2, .1, .2\]. Then the LLM will choose the index with the highest probability (index 2; remember indexing starts at 0), and then generate the corresponding token ("like"). Then based on the previous tokens ("you", and "like"), the LLM will once again generate a probability vector for the words in the vocabulary set; e.g. \[0, 0, .1, .2, .4, .3\]. Repeating the process of selecting the word indexed with the highest probability, the LLM will generate the next token ("pizza") and so on. (See the steps outlined in the Figure below.)

This is almost like a sequence of classification tasks, where each class is a possible token that can be generated.

decoding.PNG

**Temperature** is an important hyperparameter for the generation process. It determines which index the LLM will select to generate next. In the previous example, we are using a temperature of 0, which is also called "Greedy" decoding; this means that the LLM will always choose the highest probability token. If you choose a higher temperature value, the model will instead sample from the other probabilities. The higher the temperature value, the more random the sampling.

## Transformers
The most popular architecture for building LLMs are **Transformers**. The Transformer architecture is an artificial neural network that is specifically designed to deal with text data. It is composed of multiple **"self-attention"** blocks, which aggregate information between different tokens (words) in the input.

## Attention

In natural language, the meaning of a word or phrase often depends on its context – the other words around it. Self-attention (also referred to as Attention) is a mechanism that allows a model to understand the context of each word or token in a sequence by considering its relationship with all other words or tokens in that sequence.

The attention mechanism is crucial for capturing the relationships within sequences, regardless of their positions. The process involves computing three matrices: Query (Q), Key (K), and Value (V). These matrices are computed by simple Multi Layered Perceptrons. These Q, K, and V matrices are then used to compute the self-attention scores and weighted values.

SurestartFigures(8).png

These attention scores of each token represent the importance of the other tokens for the token under consideration. The self-attention mechanism computes the attention scores by taking the dot product of the query vector for each token with the key vectors of all other tokens, scaled by a factor (square root of the key dimension), and passed through a softmax function.

A calculation of self-attention scores between any two words looks like this:

SureStart figures(4).png

This is just the calculation between two tokens.

The self-attention mechanism works like this:

1. For each word or token in the sequence, the model creates three different representations called Query, Key, and Value.
2. The Query representation of a word is compared to the Key representations of all other words in the sequence. This comparison gives a score for how related or important each other word is to the current word.
3. The Value representations of the other words are then combined using the scores (computed in step 2 above) as weights. Words that are more related to the current word get higher weights, and their Value representations contribute more to the final result.
4. This weighted combination of Value representations becomes the new representation for the current word, incorporating information from the other relevant words in the sequence.

Let us consider the sentance "I like icecream today":

The self-attention mechanism allows for arbitrary length input sizes because the Q, K, and V computations, as well as the attention score calculation, can be performed in parallel for all tokens in the input sequence, regardless of the sequence length. The attention scores and weighted values are computed independently for each token, allowing the mechanism to handle sequences of varying lengths.

However, the limitation in terms of length of the input text comes from the fixed-size feed-forward neural network (FFN) layer that follows the self-attention mechanism. The FFN layer expects a fixed-size input, typically determined by the model's architecture and hyperparameters. To accommodate this constraint, the input sequence is either padded with special tokens (e.g., <pad>) or truncated to a maximum length before being passed through the FFN layer.

This limitation is similar to convolutional neural networks (CNNs) used in computer vision tasks. The convolutional layers in CNNs can process input images of arbitrary sizes by applying the convolution operation locally across the image. However, the final fully-connected layers in CNNs expect a fixed-size input, typically achieved by resizing or cropping the input images to a specific resolution.

Just like in Transformers, the convolutional layers in CNNs are not inherently limited by input size, but the fixed-size fully-connected layers introduce a constraint that requires preprocessing the input to a specific size. This trade-off between flexibility and computational efficiency is a common design choice in both Transformer and CNN architectures.

## Pretraining vs. Finetuning


**Pretraining** refers to the process of training an LLM on a lot of text to give it a general understanding of language. The developer of the LLM will collect a lot of text data, and train the model using the language modelling objective.


After pretraining, we can use various methods to improve the performance for certain tasks...

**In-Context Learning** (sometimes called ''few-shot prompting'') refers to providing a few examples to the LLM **in the prompt**, which the LLM uses as demonstrations. For example, using the previous example, if we wanted to add three in-context examples:

"Answer the question.

Q: What is the population of the United Kingdom?
A: 68 million

Q: What is the capital of Japan?
A: Tokyo

Q: What is the largest continent by land area?
A: Asia

Q: What is the capital of the United States?
A:"

**Finetuning** involves updating the parameters of the model to do specific tasks. Say, we wanted a given LLM to answer questions of a specific kind in a specific output format, like this: "Q: What is the population of the United Kingdom? A: 68 million". To finetune a pre-trained LLM model, we would start by building a training set with examples, similar in content and format. Then, we would update the model parameters using this fine-tuning training set, based on the language modelling objective.

**Instruction-tuning** is a method of finetuning that specifically teaches LLMs to follow human instructions. This improves the models' abilities to follow commands. Generally, instruction tuning improves LLM performance in dialogue applications.

## How do you pick the right LLM?
Most LLMs are made by large companies with a large amount of financial and computational resources. It is very rare that an engineer will need to build an LLM from scratch for their use case. Most of the time, the main thing is choosing the correct LLM for your task. There are a few dimensions you will want to consider.

1. API
2. Data
3. Licensing
4. Cost
5. Model Size
6. Performance

**API** - You will have to interact with different LLMs in different ways. For example, the only way to access the GPT models from OpenAI is to use their website or their web/programmatic APIs. They do not allow you to have full access to the model weights. On the other hand, a lot of Open Source models will not have nice web APIs to access the model with, often you will have to host the model yourself, and write code to run the model.

**Data** - Different models will be trained with different datasets. For example, Codex is trained with a lot of programming language data, so it is better at solving programming problems.

**Licensing** - Certain LLMs place limits on commercial use. For example, Meta requires you to request a special license for commercial use if you have more than 700M active users.

**Cost** - Different web APIs for different LLMs have different costs, and if you are running a model by yourself, you need to account for server costs.

**Model Size** - Models range between larger more capable models and smaller faster models. Larger models generally tend to stronger more descriptive learners, and produce higher quality results, but they require more time, money, and/or memory to run.

Another important factor when picking your LLM is to simply try different examples from your use case. You should make a few example use cases, and test out a few models with your example use cases. The metrics reported in a technical paper, may not accurately reflect the needs of your particular use case, so the best thing you can do is just test them yourself.
